### 진에어 국제선 사이트 크롤링
- 국제선 사이트는 .net aspx 로 개발되어 단계가 복잡함
    * 국제선 예매사이트 1차 접속 : 'https://www.jinair.com/bookflight/flightSearch.aspx'
    * 국제선 예매사이트 viewstate 값얻기 위한 2차 접속, 출발지 선택시 웹페이지에서 호출됨
    * 국제선 예매사이트 최종 검색 1단계 진행 : 다음 버튼 클릭 처리
    * 버튼 클릭시 새창으로 연결 : 'https://www.jinair.com/BookFlight/BookingProcess.aspx'
    * 새창이 열린후 다른 URL로 변경 : 'https://www.jinair.com/BookFlight/FlightSearchBridge.aspx'
        - 이때 최종 검색을 위한 파라미터를 생성 및 검색 페이지 URL 얻어옮
    * 최종 검색 정보 읽어오는 페이지를 이용 정보 검색
        - 자바스크립터로 처리 되고 있음, 필요한 JSON 부분만 발췌하여 처리

```python
def raw_to_df(raw_data):
    json_data = json.loads(raw_data)

    ## 플라이트 정보 생성
    try:
        raw_flight_info = json_data['pageDefinitionConfig']['pageData']['business']['Availability']['proposedBounds'][0]
        key_fields = ['proposedBoundId', 'beginDate','endDate','flightNumber']
    except Exception as e:
        ## 정보가 없을 경우
        print('****** No Record ******')
        return None
    flight_infos = mining_value_by_last_field_name(raw_flight_info, key_fields)

    ## 가격 정보 생성
    raw_fare_info = json_data['pageDefinitionConfig']['pageData']['business']['Availability']['recommendations']
    fare_infos = {
        'flightId':[],
        'numberOfSeatsLeft':[],
        'ffCode':[],
        'amountWithoutTax':[],
        'tax':[]
    }
    for fare_info in raw_fare_info:
        for flight_group in fare_info['bounds'][0]['flightGroupList']:
            ## 정보 추출
            fare_infos['flightId'].append(flight_group['flightId'])
            fare_infos['numberOfSeatsLeft'].append(flight_group['numberOfSeatsLeft'])
            fare_infos['ffCode'].append(fare_info['fareFamily']['ffCode'])
            fare_infos['amountWithoutTax'].append(fare_info['PNRprices'][0]['recoAmounts'][0]['recoAmount']['amountWithoutTax'])
            fare_infos['tax'].append(fare_info['PNRprices'][0]['recoAmounts'][0]['recoAmount']['tax'])

    ## proposedBoundId 와 flightId 기준으로 join
    df = pd.merge(pd.DataFrame(flight_infos),pd.DataFrame(fare_infos),left_on='proposedBoundId',right_on='flightId')
    ## 필요한 정보의 형태로 변경
    date_fmt = '%b%d,%Y%I:%M:%S%p'
    df['beginDate']=pd.to_datetime(df['beginDate'],format=date_fmt)
    df['endDate']=pd.to_datetime(df['endDate'],format=date_fmt)
    df['date'] = df['beginDate'].map(lambda x: x.strftime('%Y%m%d'))
    df['dpt'] = df['beginDate'].map(lambda x: x.strftime('%H:%M'))
    df['arr'] = df['endDate'].map(lambda x: x.strftime('%H:%M'))
    heads_map = {
        'flightNumber':'flt',
        'amountWithoutTax':'fare',
        'numberOfSeatsLeft':'seat',
    }
    data_heads = ['date','flt','dpt','arr','fare','tax','seat']
    df.rename(columns=heads_map,inplace=True)
    return df[data_heads]


## 하루치 정보 읽어오기
def read_LJ_1day_fare(dpt,arr,dpt_date):
    ## 정보 읽어오기
    df = raw_to_df(crawling_LJ_data(dpt,arr,dpt_date))
    ## 통계 정보 처리
    if df is not None:
        stat = stat_fare(df,columns=['fare'])
        df.ix[len(df)] = [dpt_date.replace('-',''),'min','max','mean',stat[0],stat[1],stat[2]]
    return df

## 정해진 기간의 데이터 읽어오기, 기본 30일
def read_LJ_date_range_fare(dpt,arr,start=0,end=31):
    # 기본 30일간 데이터 읽어서 파일로 저장
    date_range = [ (datetime.today()+timedelta(1)*i).strftime('%Y%m%d') for i in range(start,end)]
    fare_list = []
    for d in date_range:
        try:
            fare_df = read_LJ_1day_fare(dpt,arr,d)
            if fare_df is not None:
                fare_list.append(fare_df)
        except Exception as e:
            print('****** Error occured : ',e)
    result = pd.concat(fare_list,ignore_index=True)
    print('++++++++++Total : ', len(result))
    ## 파일 저장
    result.to_excel('{}/{}_{}_{}_{}_{}_{}.xls'.format('excel','LJ',dpt[:3],arr[:3],start,end,datetime.today().strftime('%Y%m%d%H%M')))
    return result
```
