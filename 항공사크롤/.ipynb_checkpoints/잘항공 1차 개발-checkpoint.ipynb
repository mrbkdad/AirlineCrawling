```python
## 모바일 사이트 이용
## http://www.kr.jal.com/krl/sp/ko/reservation/  예약 버튼 클릭
## http://www.5931.jal.co.jp/rsvInter/1AEncrypt.do 사이트를 통해 1차 처리후 리다이렉트
## https://book.jal.com/plnext/Mobile4JAL/Override.action 로 이동하여 정보 가져 오기
##    * 페이지 내에서 JSON 데이터 스크립트로 처리 - 분리후 JSON 파싱 필요
def crawl_JL(dpt,arr,dpt_date):
    url = 'http://www.kr.jal.com/krl/sp/ko/reservation/'
    head = {
        'Accept-Language':'ko-KR,ko',
        'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
    }
    sess = requests.Session()
    req = sess.get(url,headers=head)
    soup = BeautifulSoup(req.text,'lxml')
    script=soup.select('script')[17]
    start_txt = "tmSetup.SO_GL.KO = '"
    end_txt = "';\r\n\r\n"
    start_pos = script.text.find(start_txt) + len(start_txt)
    end_pos = script.text.rfind(end_txt)
    so_gl = script.text[start_pos:end_pos]

    param = {
        'TRIP_TYPE':'O',
        'B_DATE_1':dpt_date + '0000',
        'TRAVELLER_TYPE_1':'ADT',
        'LANGUAGE':'KO',
        'B_LOCATION_1':dpt,
        'E_LOCATION_1':arr,
        'SO_SITE_EXTERNAL_ID':'Mobile4JAL',
        'COMMERCIAL_FARE_FAMILY_1':'3KRECOOW',
        'DATE_RANGE_VALUE_1':'0',
        'DATE_RANGE_VALUE_2':'0',
        'FIELD_ADT_NUMBER':'1',
        'FIELD_CHD_NUMBER':'0',
        'FIELD_INFANTS_NUMBER':'0',
        'SO_SITE_ENABLE_ERR_REDIRECT':'TRUE',
        'UI_EMBEDDED_TRANSACTION':'MAvailabilityFlowDispatcher',
        'SEARCH_PAGE':'FP',
        'MT':'A',
        'SO_SITE_OFFICE_ID':'SELJL18BB',
        'SO_SITE_QUEUE_OFFICE_ID':'SELJL1200',
        'SO_SITE_QUEUE_CATEGORY':'1C1',
        'SO_SITE_POINT_OF_SALE':'SEL',
        'SO_SITE_POINT_OF_TICKETING':'SEL',
        'SO_SITE_FD_DISPLAY_MODE':'1',
        'SO_SITE_ETKT_Q_OFFICE_ID':'SELJL1200',
        'SO_SITE_ETKT_Q_AND_CAT':'8C1',
        'SO_GL':so_gl,
        'SO_SITE_MAIL_FROM':'Japan Airlines <org.selsst.jali@jal.com>',
        'SO_SITE_PNR_ADD_EMAIL1':'Japan Airlines <org.selsst.jali@jal.com>',
        'SITE':'C002C002',
        'SO_SITE_MANUAL_ETKT_CMD':'TTP/T-JL/ITR-EMLA/LA-KO',
        'SO_SITE_CURRENCY_FORMAT_JAVA':'0',
        'SO_SITE_MC_ENBL_FF_URI_POPUP':'TRUE',
        'ENCODE_CHARSET':'EUC-KR',
        'REDIRECT_URL':'https://book.jal.com/plnext/Mobile4JAL/Override.action',
        'SO_SITE_MC_ALLOW_SEAT_BOOK':'FALSE',
        'SO_LANG_HOMEPAGE_URL':'http://www.jal.com/',
    }
    url = 'http://www.5931.jal.co.jp/rsvInter/1AEncrypt.do'
    head = {
        'Accept-Language':'ko-KR,ko',
        'Referer':'http://www.kr.jal.com/krl/sp/ko/reservation/',
        'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
    }
    req = sess.post(url,param,headers=head)
    soup = BeautifulSoup(req.text,'lxml')
    param = {input['name']:input['value'] for input in soup.select('input')}
    url = 'https://book.jal.com/plnext/Mobile4JAL/Override.action'
    head = {
        'Accept-Language':'ko-KR,ko',
        'Referer':'http://www.5931.jal.co.jp/rsvInter/1AEncrypt.do',
        'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
    }
    req = sess.post(url,param,headers=head)
    raw_txt = req.text
    start_txt = 'json:'
    end_txt = '}'
    start_pos = raw_txt.find(start_txt) + len(start_txt)
    end_pos = raw_txt.rfind(end_txt)
    return raw_txt[start_pos:end_pos].strip()
## 포맷 : 대상항공사코드,플라이트,출발일,출발지,도착지,출발시간,도착시간,가격,유류세,세금,좌석수
def scraping_JL(raw_data):
    raw_json = json.loads(raw_data)
    if 'MOWDFFSORT_A' not in raw_json['data']['booking'].keys():
        return None
    raw_json = json.loads(raw_json['data']['booking']['MOWDFFSORT_A']['requestParam']['fpowPanelKey']['generatedJSonByPrice'])
    fare_list = []
    for flight in raw_json['list_tab']['list_proposed_bound'][0]['list_flight']:
        for segment in flight['list_segment']:
            for price in flight['list_price']:
                line_data = [segment['airline']['code'],segment['flight_number'],segment['b_date_date'],
                    segment['b_location']['location_code'],segment['e_location']['location_code'],
                    segment['b_date_time'],segment['e_date_time']]
                line_data.append(price['totalPrice'])
                line_data.extend([0,0,9]) ## 유류세, 공항세, 좌석수는 임의로 처리
                fare_list.append(line_data)
    return fare_list
```
